Quick program to try to analyze what's the resolution of the accelerometer.

Reminder: Data is in m/s^2 for tdcsfog and g for defog and notype

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import os
```

```python
# Import defog
path = "dataset/train/defog"

AccML_values = set()

for file in os.listdir(path):
    df = pd.read_csv( os.path.join(path,file) )
    
    AccML_values.update( set(df.AccML) )
    

arr = np.array(list(AccML_values))    # Convert set to ndarray
arr.sort()
diff = np.diff(arr, axis=0) # Get new ndarray where each row is the difference between 2 successive rows of arr
diff
```

```python
x = np.arange(len(diff))
y = np.array(diff)
plt.plot(x, y)
```
