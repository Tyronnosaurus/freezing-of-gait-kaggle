# Data cleaning

Defog constraints:
- Time: must start from 0, always increase by 1 (before dropping any rows)
- AccV, AccML, AccAP: check for suspicious zeroes, ridiculously high values, NaNs.
- StartHesitation, Turn, Walking: Only 0 or 1. Only one 1 per row.
- Valid: Only true or false.
- Task: Only true or false.

```python
# For now we check first dataframe. TODO: do it for all of them
df = defog_dfs[0]
df.set_index("Time")
```

```python
# Time: 
assert(df.Time[0] == 0)                 # Must start from 0
assert(df.Time.is_monotonic_increasing) # Must always increase by 1 (before we drop any rows)

# AccV, AccML, AccAP:
assert(0 not in df.AccV.values)
assert(0 not in df.AccML.values)
assert(0 not in df.AccAP.values)
```
